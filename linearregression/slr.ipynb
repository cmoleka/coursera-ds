# Import Dataset

Importing a dataset to practice with. Doing so will allow me to get used to a Data working pattern.


```python
file_name='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
df=pd.read_csv(file_name)
```

## Dataset description and missing values analysis
We get a statistical summary of each column and evaluate missing values.

```python
# Show data types of each columns
df.dtypes
```

```python
# Get statistical summary of each column.
df.describe()
```

Get number of missing values and Display missing values as row of the data frame.

```python
# Find missing values
df.isna().sum()
```

```python
df[df.isnull().any(axis=1)].head(5)
```

We can replace the missing values of bedrooms and bathrooms with the mean of each columns.

```python
bedroom_mean = df['bedrooms'].mean()
bathrooms_mean = df['bathrooms'].mean()
df['bedrooms'].replace( np.nan, bedroom_mean, inplace=True)
df['bathrooms'].replace( np.nan, bathrooms_mean, inplace=True)
```

```python
# drop useless columns
df.drop(['Unnamed: 0', 'id'], axis=1, inplace=True)
df.describe()
```
