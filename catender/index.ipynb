```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tpsgc.csv')
df2 = df.dropna(subset=['non_competitive_procurement_strategy'])
df2 = df2.drop(['region_opportunity', 'amendment_date','competitive_procurement_strategy', 'document', 'attachment'], axis=1)
df2 = df2.replace('Français', 'French')
df2 = df2.replace('Attribué', 'Awarded')
df2 = df2.replace('Expiré', 'Expired')
df2 = df2.replace('Annulé', 'Cancelled')
df2 = df2.replace('Actif', 'Active')
df2.reset_index(drop=True, inplace=True)
df2.head()
```
